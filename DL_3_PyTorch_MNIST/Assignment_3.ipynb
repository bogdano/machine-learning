**Problem 3 (40 POINTS)**.  In this problem we will start by creating the most naive architecture to solve an image recognition problem, in which we use only traditional fully-connected layers. **Part (a)**. Create a fully-connected neural network with 8 total layers of parameters.  Aside from the output layer, each layer should have 50 hidden units.  Use ReLU activations.  We will call this network *NetFC*.  

```python
class NetFc(nn.Module):
    def __init__(self):
      super().__init__()      
      # layers of the network, 8 layers w 50 hidden units
      self.fc1 = nn.Linear(28*28, 50) # 28*28 is the dimension of an MNIST image
      self.fc2 = nn.Linear(50, 50)
      self.fc3 = nn.Linear(50, 50)
      self.fc4 = nn.Linear(50, 50)
      self.fc5 = nn.Linear(50, 50)
      self.fc6 = nn.Linear(50, 50)
      self.fc7 = nn.Linear(50, 50)
      self.fc8 = nn.Linear(50, 10) # output maps to 10 classes

    def forward(self, x):
      # forward pass through the network
      x = x.view(-1, 28*28)
      x = F.relu(self.fc1(x))
      x = F.relu(self.fc2(x))
      x = F.relu(self.fc3(x))
      x = F.relu(self.fc4(x))
      x = F.relu(self.fc5(x))
      x = F.relu(self.fc6(x))
      x = F.relu(self.fc7(x))
      # no relu on output layer
      output = self.fc8(x)
      # softmax layer should always be last
      # output = F.log_softmax(x, dim=1)
      return output
```

**Part (b)**. In this part you will create a function that trains your model for two epochs (i.e., two complete passes over the training data), and prints the loss of your model on each 100th minibatch. Use the Adam optimizer, and a cross entropy loss for classification.

```python

# This function takes in a pytorch model as 'net', and then trains it for two
# epochs (i.e., two loops over teh entire dataset).  You may want to adjust
# the learning rate sometimes and we will pass that in as 'lr'.  You will also
# later need to vary the number of epochs of training, so we will pass that in
# as 'n_epochs.  You will also need to pass in the 'trainloader' object to
# handle your data. Your function should return the trained model.

def trainMyModel(net, lr, trainloader, n_epochs):
  # Attempt to put your neural network onto the GPU. Do not worry if there is no
  # GPU available on CoLab; for models of our size it won't be prohibitively
  # slow and you can proceed with a CPU
  device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
  print(f"Training on {device}")
  net.to(device)
  for epoch in range(n_epochs):
    # loss function
    criterion = nn.CrossEntropyLoss()
    # optimizer
    optimizer = optim.Adam(net.parameters(), lr=lr)
    # store running loss
    running_loss = 0.0
    
    for i, data in enumerate(trainloader, 0):
      inputs, labels = data[0].to(device), data[1].to(device)
      optimizer.zero_grad()
      # forward + backward + optimize
      outputs = net(inputs)
      loss = criterion(outputs, labels)
      loss.backward()
      optimizer.step()
      # print statistics
      running_loss += loss.item()
      if i % 100 == 99:    # print every 100 mini-batches
        print('[epoch: %d, batch: %3d] loss: %.3f' % (epoch + 1, i + 1, running_loss / 100))
        running_loss = 0.0

  print('✨ Finished Training ✨')
  return net
```

**Part (c)**. In this part you will create a function that will take a trained model as input, and a data loader (for your test data), and return the accuracy of the model on the testing data.  The accuracy should be reported to at least two decimal places. This function need not be general, but it should work with the 'testloader' that was created above for the MNIST dataset.

```python
# This function takes in a pytorch model as 'trainedNet', and then applies it
# to testing data that can be loaded using the 'testloader' object.  The
# function returns the accuracy of trainedNet on the whole testing dataset
# and prints the model's accuracy to the terminal (to at least two decimal places of accuracy)
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
def testMyModel(trainedNet, testloader):
  correct = 0
  total = 0
  with torch.no_grad():
    for data in testloader:
      images, labels = data[0].to(device), data[1].to(device)
      outputs = trainedNet(images)
      _, predicted = torch.max(outputs.data, 1)
      total += labels.size(0)
      correct += (predicted == labels).sum().item()
  acc = 100 * correct / total
  
  print(f'Accuracy of the network on the {len(testloader.dataset)} test images: {acc} %')
  return acc
```

**Problem 4 (10 POINTS)**.  Now we will see the advantages of skip connections in a deep neural network.  Below, fill in the template to create another fully connected neural network, called *NetFcRes* that has two "skip connections" in it.  By "skip connection", this means a direct connection from an earlier layer to a later layer in the network.   For example, a skip connection between the first layer and third layer means that the input to the fourth layer is the sum of the output of first layer and the third layers. In this problem, create a network with two skip connections: one from the output of the first layer to the output of the third layer; and one from the output of the fifth layer to  the output of the 7th layer.  Then use your training and testing functions to train and evaluate the performance of your *NetFcRes* model.  

Adding skip connections results in so-called "residual networks", which are now widely-used in research and practice.  We have not discussed these yet in class, but simply follow the instructions above.  If you follow the instructions above you will only need to make very minor modifications to *NetFc* to obtain your new *NetFcRes* model.  If you are interested however, or prefer to read more, search for "residual layer" or "skip connection" on google; there are substantial resources explaining this structure. This topic is also discussed in Chapter 11 of our course textbook.  

If done correctly, this simple change should consistently improve your model's performance, and if done correctly, you should now obtain at least 93% accuracy on the testing set after training two epochs.

```python
class NetFcRes(nn.Module):
    def __init__(self):
      super().__init__()      
      # layers of the network, 8 layers w 50 hidden units
      self.fc1 = nn.Linear(28*28, 50)
      self.fc2 = nn.Linear(50, 50)
      self.fc3 = nn.Linear(50, 50)
      self.fc4 = nn.Linear(50, 50)
      self.fc5 = nn.Linear(50, 50)
      self.fc6 = nn.Linear(50, 50)
      self.fc7 = nn.Linear(50, 50)
      self.fc8 = nn.Linear(50, 10)

    def forward(self, x):
      # forward pass through the network
      x = x.view(-1, 28*28)
      x1 = F.relu(self.fc1(x))
      x2 = F.relu(self.fc2(x1))
      # add a lil skip from layer 1
      x3 = F.relu(self.fc3(x2) + x1)
      x4 = F.relu(self.fc4(x3))
      x5 = F.relu(self.fc5(x4))
      x6 = F.relu(self.fc6(x5))
      # add a lil skip from layer 5
      x7 = F.relu(self.fc7(x6) + x5)
      output = self.fc8(x7)
      # softmax layer should always be last
      # output = F.log_softmax(x8, dim=1)
      return output
```

```python
# Train your model.
net = NetFcRes();
lr = 0.01;
n_epochs = 2;
trainedNet = trainMyModel(net,lr,trainloader,n_epochs);

# Test your model
testMyModel(trainedNet,testloader)
```

(a) Please create a function called "calculate_error" that takes a model and a data_loader and returns the loss on the dataset in the data loader. A code skeleton is provided below.   

**ANSWER:**

```python
#Function to compute error for a given input dataset and model
def calculate_error(model, data_loader):
  device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
  model.to(device)
  # set model to evaluation mode
  model.eval()
  criterion = nn.CrossEntropyLoss()
  error = 0.0
  with torch.no_grad():
    for data in data_loader:
      inputs, labels = data[0].to(device), data[1].to(device)
      outputs = model(inputs)
      loss = criterion(outputs, labels)
      error += loss.item()
  return error / len(data_loader)
```

(c) Create a new model called "NetFcResDropout" and add one layer of dropout to the very first layer of weights, with a dropout rate of 20%.  Again train for 40 epochs with a learning rate of 0.01.  Use your "calculate_error" function to compute the loss of the trained model on both the training and the testing set.  How do the losses now compare to those without dropout?  Briefly explain why the relationship of the training and testing loss have changed.

**ANSWER:**

```python
class NetFcResDropout(nn.Module):
    def __init__(self):
      super().__init__()
      # layers of the network, 8 layers w 50 hidden units
      self.fc1 = nn.Linear(28*28, 50)
      self.fc2 = nn.Linear(50, 50)
      self.fc3 = nn.Linear(50, 50)
      self.fc4 = nn.Linear(50, 50)
      self.fc5 = nn.Linear(50, 50)
      self.fc6 = nn.Linear(50, 50)
      self.fc7 = nn.Linear(50, 50)
      self.fc8 = nn.Linear(50, 10)
      # dropout layer with 20% dropout rate
      self.dropout = nn.Dropout(p=0.2)
      
    def forward(self, x):
      # forward pass through the network
      x = x.view(-1, 28*28)
      x1 = F.relu(self.fc1(x))
      x1 = self.dropout(x1)
      x2 = F.relu(self.fc2(x1))
      # add a lil skip from layer 1
      x3 = F.relu(self.fc3(x2) + x1)
      x4 = F.relu(self.fc4(x3))
      x5 = F.relu(self.fc5(x4))
      x6 = F.relu(self.fc6(x5))
      # add a lil skip from layer 5
      x7 = F.relu(self.fc7(x6) + x5)
      output = self.fc8(x7)
      # softmax layer should always be last
      # output = F.log_softmax(x8, dim=1)

      # Return the output of the network
      return output
```

```python
# Train your model.
# net = NetCnn();
net = NetFcResDropout()
lr = 0.01;
n_epochs = 40;
trainedNet = trainMyModel(net,lr,trainloader_small,n_epochs);

# Test your model
testMyModel(trainedNet,testloader)
```

```python
# Report training error and testing error
training_error_no_reg = calculate_error(trainedNet, trainloader_small)
testing_error_no_reg = calculate_error(trainedNet, testloader)
print(f"Training error: {training_error_no_reg:.3f}")
print(f"Testing error:  {testing_error_no_reg:.3f}")
```

*The losses are noticeably lower with a dropout layer added to the model, although this is not so clearly reflected in the accuracy of predictions. Dropout is a regularization "hack" in which the output of a chosen subset of hidden units is clamped down to zero, and the network is as such encouraged to have weights with smaller magnitudes. In this particular case, considering the fact that the dataset is tiny, it would be desirable to have the model decrease its amount of overfitting to each datapoint.*
